```python
import pandas as pd
df = pd.read_csv("http://cee.eng.usf.edu/faculty/flm/CGN6933/TTE6307-MID(21).txt",
delimiter = "\t", header=None
)
dxf = df.iloc[:,[24,22, 29, 30, 20]]
dxf.columns = ['Age', 'Gender', 'Home', 'household', 'Use_of_autonomous_vehicles']
```

```python
dxf
```

```python
# Task: Prepare a descriptive statistics, crosstab, and figures of the given data
```

```python
# checking for any null values before performing a statisstical analysis
dxf.info()
```

```python
# Preparing a descriptive statistics
#  A statistical summary the dataframe to show how data is distributed; Mean, S.D, min-max and percentiles.

dxf.describe()
```

```python
# Preparing a crosstab for feature analysis

# Relationship between age and use of autonomous vehicles
age_crosstab = pd.crosstab([dxf.Age], dxf.Use_of_autonomous_vehicles,  margins=True)
print(age_crosstab)
```

```python
# Relationship between household income and use of autonomous vehicles
income_crosstab = pd.crosstab([dxf.household], dxf.Use_of_autonomous_vehicles,  margins=True)
print(income_crosstab)
```

```python
# Relationship between gender and use of autonomous vehicles
gender_crosstab = pd.crosstab([dxf.Gender], dxf.Use_of_autonomous_vehicles,  margins=True)
print(gender_crosstab)
```

```python
# Preparing figures for the data

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting a correlation matrix between variables

correlation_mat = dxf.corr()
sns.heatmap(correlation_mat, annot = True)
plt.show()
print("Figure 1: A heatmap showing correlation between all variables")
```

```python
# Plotting a scatter chart to show gneral relationship between age, household earnings against the adoption of autonomous vehicles

df = pd.DataFrame(dxf, columns=['Age', 'household', 'Use_of_autonomous_vehicles'])
ax1 = df.plot.scatter(x='Age', y='household', c='Use_of_autonomous_vehicles', colormap='viridis')
print("Figure 2: A scatter chart showing relationship between age, household earning and adoption of auto. vehicles")
```
